```python
cur = conn.cursor()
```

```python
cur.execute("SELECT * FROM edges ORDER BY _area_id")
```

```python
sql = "UPDATE edges SET area = %s WHERE edges._area_id = %s"
```

```python
for row in cur:
    cur1 = conn.cursor()
    print(row[1])
    print(cur.rownumber)
    area_id = row[1]
    count = cur.rownumber
    print(cur.mogrify(sql, (count, area_id)))
    cur1.execute(sql, (count, area_id))
    print(cur)
#     conn.commit()
```
